# Data Cleaning for Modeling

```python
employee_df.head(3)
```

```python
employee_categorical = employee_df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']]
```

```python
employee_categorical.head()
```

```python
#Importing categorical pre-processing library
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
```

```python
employee_categorical = onehotencoder.fit_transform(employee_categorical).toarray()
```

```python
employee_categorical.shape
```

```python
employee_categorical= pd.DataFrame(employee_categorical)
employee_categorical.head()
```

```python
employee_numeric = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
employee_numeric
```

```python
employee_modeling = pd.concat([employee_categorical,employee_numeric], axis =1)
```

```python
employee_modeling.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X = scaler.fit_transform(employee_modeling)
```

```python
X
```

```python
y = employee_df['Attrition']
y
```

# Train Test Split

```python
#Import train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Model Selection

```python
#Importing model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
logRegModel= LogisticRegression()
logRegModel.fit(X_train, y_train)
```

```python
y_pred = logRegModel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print('Accuracy {} %'.format(100* accuracy_score(y_pred, y_test)))
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rndforest_model= RandomForestClassifier()
```

```python
rndforest_model.fit(X_train,y_train)
```

```python
y_pred_rf=rndforest_model.predict(X_test)
```

```python
print('Accuracy {} %'.format(100* accuracy_score(y_pred_rf, y_test)))
```

```python
print(classification_report(y_test, y_pred_rf))
```
